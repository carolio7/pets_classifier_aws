# 4. Présentation des résultats

## 4.a) Résultats obtenues

### I - Performance One-versus-All

```python
import sys, os
from math import ceil
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df_ovr = pd.read_csv('./output/result_ONE_VS_ALL_10_split.csv', index_col=[0])
df_ovr
```

### Statistique OVA:

```python
df_ovr_transposed = df_ovr.transpose()
df_ovr_transposed.describe()
```

## Performance des modèles OVA :

```python
ncolumn= 3
nrow = ceil(len(df_ovr_transposed.columns) / ncolumn)
fig, axs = plt.subplots(nrows=nrow, ncols=ncolumn,
                        figsize=(ncolumn*6, nrow*3.5), 
                        constrained_layout=True)
axsFlat = axs.flat
index = 0
axe = df_ovr_transposed.index

for pet in df_ovr_transposed.columns:
    axsFlat[index].plot(axe, df_ovr_transposed[pet])
    axsFlat[index].set_xlabel('Training data', fontsize=8)
    axsFlat[index].set_ylabel('Precision', fontsize=14)
    axsFlat[index].set_title(pet + '-vs-All', fontsize=24)
    
    index += 1

#fig.savefig('One_vs_All.png', bbox_inches='tight')
```

# Courbe de performance des modèles OVO

```python
def sauver(filenom, df, indice):
    ncolumn= 3
    nrow = 10
    fig, axs = plt.subplots(nrows=nrow, ncols=ncolumn,
                            figsize=(ncolumn*6, nrow*3.5), 
                            constrained_layout=True)
    
    axsFlat = axs.flat
    index = 0
    liste = df.columns
    while index < 30:
    #for pet in df_ovo_transposed.columns:
        pet = liste[index + indice]
        df_cutted = df[[pet]].dropna()
        axsFlat[index].plot(df_cutted.index, df_cutted[pet])
        axsFlat[index].set_xlabel('Training data', fontsize=10)
        axsFlat[index].set_ylabel('Precision', fontsize=10)
        axsFlat[index].set_title(pet, fontsize=16)
        index += 1
    #fig.savefig(filenom, bbox_inches='tight')
```

```python
sauver('./OVO/One_vs_One1.png', df_ovo_transposed, 0)
```

```python
sauver('./OVO/One_vs_One2.png', df_ovo_transposed, 30)
```
